```python
import cv2 as cv
from matplotlib import pyplot as plt

from pipeline_functions import straighten_page, remove_grid_lines
```

Load image

```python
image = cv.imread('data/img_20.jpg', cv.IMREAD_GRAYSCALE)

plt.figure(figsize=(20, 20))
plt.imshow(image, cmap='gray')
plt.show()
```

Straighten image

```python
image_straightened = straighten_page(image)

plt.figure(figsize=(20, 20))
plt.imshow(image_straightened, cmap='gray')
plt.show()
```

Remove grid lines

```python
image_no_grid = remove_grid_lines(image_straightened)

plt.figure(figsize=(20, 20))
plt.imshow(image_no_grid, cmap='gray')
plt.show()
```
